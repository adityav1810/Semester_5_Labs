```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
image_path='/content/lab9.jpg'
```

```
im=cv2.imread(image_path)
```

```
cv2_imshow(im)
print(im.shape)
```

```
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
gradient = cv2.morphologyEx(im, cv2.MORPH_GRADIENT, kernel) 
```

```
cv2_imshow(gradient)
```
