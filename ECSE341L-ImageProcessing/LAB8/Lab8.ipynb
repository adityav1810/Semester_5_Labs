```python
import cv2
from google.colab.patches import cv2_imshow


fingerprint_path='/content/Fingerprint.png'
mandril_path='/content/mandrill_noisy.png'
originalImage = cv2.imread(mandril_path)
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
  
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
print('Black white image')
cv2_imshow(blackAndWhiteImage)
print('Original Image')
cv2_imshow(originalImage)
print('Gray Image')
cv2_imshow( grayImage)
  
cv2.imwrite('bwimg.jpg',blackAndWhiteImage)    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Q1

```python
import matplotlib.pyplot as plt 

from skimage.io import imread, imsave 

from skimage.morphology import erosion, dilation, opening, closing, white_tophat 

from skimage.morphology import black_tophat, skeletonize, convex_hull_image 

from skimage.morphology import disk, square, diamond 

from skimage.color import rgb2gray 

from skimage.util import img_as_ubyte 
bwimg_path='/content/bwimg.jpg'
im = imread(bwimg_path, as_gray=True) 
im = rgb2gray(im) 
plt.imshow(im, cmap='gray') 
plt.show() 


selem = disk(6)


eroded = erosion(im, selem)

dilated = dilation(im, selem)

opened = opening(im, selem)

closed = closing(im, selem)


i=0
plt.figure(figsize=(9,7))
plt.subplot(2,2,i+1), plt.imshow(eroded, cmap='gray'),plt.title('Erosion')

plt.subplot(2,2,i+2), plt.imshow(dilated, cmap='gray'),plt.title('Dilation')

plt.subplot(2,2,i+3), plt.imshow(opened, cmap='gray'),plt.title('Opening')

plt.subplot(2,2,i+4), plt.imshow(closed, cmap='gray'),plt.title('Closing')

plt.show()
```

```python
img = imread(mandril_path, as_gray=True) 
img = rgb2gray(img) 

selem = square(2)
closed = closing(img, selem)
opened = opening(img,selem)
both = closing(opened,selem)

i=0
plt.figure(figsize=(9,7))

plt.subplot(1,2,i+1), plt.imshow(img, cmap='gray'),plt.title('Original')

plt.subplot(1,2,i+2), plt.imshow(opened, cmap='gray'),plt.title('Opening')

plt.show()

i=0
plt.figure(figsize=(9,7))

plt.subplot(1,2,i+1), plt.imshow(img, cmap='gray'),plt.title('Original')

plt.subplot(1,2,i+2), plt.imshow(closed, cmap='gray'),plt.title('Closing')

plt.show()

i=0
plt.figure(figsize=(9,7))

plt.subplot(1,2,i+1), plt.imshow(img, cmap='gray'),plt.title('Original')

plt.subplot(1,2,i+2), plt.imshow(both, cmap='gray'),plt.title('Opening then Closing')

plt.show()
```
